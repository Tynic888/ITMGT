```python
import requests 
import matplotlib.pyplot as plt
import time

x_values = []
y_values = []

x_max = 180
y_max = 90
x_min = -180
y_min = -90

plt.title("ISS Locator")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.plot(x_max, y_max)
plt.plot(x_min, y_min)

time = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
timestamp = time.json()
print("Start:" + str(timestamp.get("timestamp")))

for i in range(0,100):
    response = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
    time.sleep(10)
    if response.status_code == 200:
        data = response.json()
        latitude = data.get("latitude")
        longitude = data.get("longitude")
        x_values.append(longitude)
        y_values.append(latitude)

time_2 = requests.get("https://api.wheretheiss.at/v1/satellites/25544")
timestamp_2 = time_2.json()
print("End:" + str(timestamp_2.get("timestamp")))

plt.plot(x_values, y_values)
plt.show()
```
